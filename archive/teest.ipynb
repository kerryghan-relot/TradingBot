```python
# Recupere le fichier resultats_backtest.csv et affiche les Buy&Hold % les plus eleves
import pandas as pd

# Lire le fichier CSV
df = pd.read_csv('../resultats/resultats_backtest.csv')

# Afficher un tableau avec les Buy&Hold % les plus eleves
if 'Buy&Hold %' in df.columns:
    top_bh = df.sort_values(by='Buy&Hold %', ascending=False)
    display(top_bh[['Symbole', 'Stratégie', 'Buy&Hold %']].head(10))
else:
    print("La colonne 'Buy&Hold %' n'est pas presente dans le fichier CSV.")
```
